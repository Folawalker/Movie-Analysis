# Correlation Analysis

```python
plt.title("Budget vs Gross Earnings")
sns.regplot(x="Gross", y="Budget", data=movies, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
plt.show()
```

```python
numeric_movies = movies.select_dtypes(include='number')
```

```python
correlation =numeric_movies.corr(method='pearson')
numeric_movies.corr(method='pearson')
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(correlation, annot=True, cmap='coolwarm_r')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

# Critical Acclaim of Directors

## Most Popular Directers By Number of Movies

```python
movies.groupby('Director').size().reset_index(name='Movie Count').sort_values(by='Movie Count', ascending=False).head(10)
```

## By Rating

```python
movies.groupby('Director', as_index=False)['Score'].mean().round(1).sort_values(by='Score', ascending=False).head(10)
```

```python
movies.groupby('Director', as_index=False)['Score'].mean().round(1).sort_values(by='Score', ascending=True).head(10)
```

## Directors With Highest Peak

```python
movies.groupby('Director', as_index=False)['Score'].max().round(1).sort_values(by='Score', ascending=False).head(10)
```

## By Budget

```python
movies.groupby('Director', as_index=False).agg(
    Movie_Count=('Name', 'count'),  # Count number of movies per director
    Total_Budget=('Budget', 'sum')   # Sum total budget per director
).sort_values(by='Total_Budget', ascending=False).head(10)
```

```python
movies.groupby('Director', as_index=False).agg(
    Movie_Count=('Name', 'count'),  # Count number of movies per director
    Total_Budget=('Budget', 'sum')   # Sum total budget per director
).sort_values(by='Total_Budget', ascending=True).head(10)
```

## By Revenue

```python
movies.groupby('Director', as_index=False).agg(
    Movie_Count=('Name', 'count'),  # Count number of movies per director
    Total_Revenue=('Gross', 'sum')   # Sum total budget per director
).sort_values(by='Total_Revenue', ascending=False).head(10)
```

```python
movies.groupby('Director', as_index=False).agg(
    Movie_Count=('Name', 'count'),  # Count number of movies per director
    Total_Revenue=('Gross', 'sum')   # Sum total budget per director
).sort_values(by='Total_Revenue', ascending=True).head(10)
```
